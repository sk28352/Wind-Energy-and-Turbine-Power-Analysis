```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from scipy import stats, special
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load the dataset
file_path = 'wind_dataset_(2).csv'
df = pd.read_csv(file_path)

# Preprocess data
df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)
y = df['WIND'].values

# Train-test split
train_size = int(len(y) * 0.8)
train, test = y[:train_size], y[train_size:]

# Box-Cox Transformation
train_boxcox, lam = stats.boxcox(train + 1e-8)
test_boxcox = stats.boxcox(test + 1e-8, lmbda=lam)

# Log Transformation
train_log = np.log(train + 1e-8)
test_log = np.log(test + 1e-8)

# Function to calculate SMAPE and MAPE
def smape(y_true, y_pred):
    return 100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))

def mape(y_true, y_pred):
    return 100 * np.mean(np.abs((y_true - y_pred) / y_true))

# Fine-tuning function
def fine_tune_arima(train, test, order, transformation, lam=None):
    model = ARIMA(train, order=order)
    model_fit = model.fit()
    predictions = model_fit.forecast(steps=len(test))
    
    if transformation == 'boxcox':
        predictions = special.inv_boxcox(predictions, lam)
    elif transformation == 'log':
        predictions = np.exp(predictions)
        
    return predictions, model_fit

def fine_tune_sarima(train, test, order, seasonal_order, transformation, lam=None):
    model = SARIMAX(train, order=order, seasonal_order=seasonal_order)
    model_fit = model.fit(disp=False)
    predictions = model_fit.forecast(steps=len(test))
    
    if transformation == 'boxcox':
        predictions = special.inv_boxcox(predictions, lam)
    elif transformation == 'log':
        predictions = np.exp(predictions)
        
    return predictions, model_fit

# Define orders for ARIMA and SARIMA
arima_order = (2, 0, 2)
sarima_order = (2, 0, 2)
seasonal_order = (1, 1, 1, 12)

# Evaluate ARIMA with Box-Cox
pred_arima_boxcox, model_arima_boxcox = fine_tune_arima(train_boxcox, test_boxcox, arima_order, 'boxcox', lam)
evaluate_model(test, pred_arima_boxcox, "Fine-Tuned ARIMA with Box-Cox Transformation")

# Evaluate SARIMA with Box-Cox
pred_sarima_boxcox, model_sarima_boxcox = fine_tune_sarima(train_boxcox, test_boxcox, sarima_order, seasonal_order, 'boxcox', lam)
evaluate_model(test, pred_sarima_boxcox, "Fine-Tuned SARIMA with Box-Cox Transformation")

# Evaluate ARIMA with Log
pred_arima_log, model_arima_log = fine_tune_arima(train_log, test_log, arima_order, 'log')
evaluate_model(test, pred_arima_log, "Fine-Tuned ARIMA with Log Transformation")

# Evaluate SARIMA with Log
pred_sarima_log, model_sarima_log = fine_tune_sarima(train_log, test_log, sarima_order, seasonal_order, 'log')
evaluate_model(test, pred_sarima_log, "Fine-Tuned SARIMA with Log Transformation")

# Plotting Training, Validation, and Residuals for SARIMA with Log Transformation
plt.figure(figsize=(14, 7))

# Training vs Prediction
plt.subplot(3, 1, 1)
plt.plot(train, label='Training Data')
plt.plot(train_size + np.arange(len(test)), test, label='Test Data', color='orange')
plt.plot(train_size + np.arange(len(test)), pred_sarima_log, label='SARIMA Log Predictions', color='green')
plt.title('Training and SARIMA Log Prediction')
plt.legend()

# Validation vs Prediction
plt.subplot(3, 1, 2)
plt.plot(test, label='Actual Test Data', color='orange')
plt.plot(pred_sarima_log, label='SARIMA Log Predicted Data', color='green')
plt.title('SARIMA Log Validation and Prediction')
plt.legend()

# Residuals Plot
residuals_sarima_log = test - pred_sarima_log
plt.subplot(3, 1, 3)
plt.plot(residuals_sarima_log, label='SARIMA Log Residuals', color='red')
plt.title('SARIMA Log Residuals')
plt.legend()

plt.tight_layout()
plt.show()
```
