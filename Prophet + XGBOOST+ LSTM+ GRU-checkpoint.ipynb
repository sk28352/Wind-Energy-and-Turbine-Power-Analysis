```python
import pandas as pd
import numpy as np
from prophet import Prophet
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Function to calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Function to calculate RMSE
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Function to calculate SMAPE
def symmetric_mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred))) * 100

# Load and preprocess the dataset
file_path = 'wind_dataset_(2).csv'
df = pd.read_csv(file_path)

# Preprocess data for Prophet
df['DATE'] = pd.to_datetime(df['DATE'])
df = df.rename(columns={'DATE': 'ds', 'WIND': 'y'})  # Rename columns for Prophet
df = df[['ds', 'y']]  # Select only the relevant columns

# Split data into train and test
train_size = int(len(df) * 0.8)
train_df, test_df = df[:train_size], df[train_size:]

# Train Prophet model
model = Prophet()
model.fit(train_df)

# Make future dataframe and predictions
future = model.make_future_dataframe(periods=len(test_df), freq='D')
forecast = model.predict(future)

# Extract the forecasted values and calculate residuals
y_pred_prophet = forecast['yhat'].iloc[-len(test_df):].values
y_true = test_df['y'].values

# Fine-Tuning XGBoost
xgb_model = XGBRegressor(colsample_bytree=0.788992592521131, learning_rate=0.05968705502967452, max_depth=2, n_estimators=64, subsample=0.8268511842933991)
xgb_model.fit(test_df.index.values.reshape(-1, 1), y_true - y_pred_prophet)
residuals_pred_xgb = xgb_model.predict(test_df.index.values.reshape(-1, 1))
final_pred_xgb = y_pred_prophet + residuals_pred_xgb

# Fine-Tuning Gradient Boosting
gb_model = GradientBoostingRegressor(learning_rate=0.011185556452886727, max_depth=4, n_estimators=121, subsample=0.9045585190672509)
gb_model.fit(test_df.index.values.reshape(-1, 1), y_true - y_pred_prophet)
residuals_pred_gb = gb_model.predict(test_df.index.values.reshape(-1, 1))
final_pred_gb = y_pred_prophet + residuals_pred_gb

# LSTM Model
scaler = MinMaxScaler(feature_range=(0, 1))
residuals_scaled = scaler.fit_transform((y_true - y_pred_prophet).reshape(-1, 1))
look_back = 30
train_residuals_scaled = residuals_scaled[:train_size - look_back]
train_generator = TimeseriesGenerator(train_residuals_scaled, train_residuals_scaled, length=look_back, batch_size=32)

lstm_model = Sequential()
lstm_model.add(LSTM(50, activation='relu', input_shape=(look_back, 1)))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(train_generator, epochs=20, verbose=0)

lstm_predictions = []
first_eval_batch = residuals_scaled[-look_back:]
current_batch = first_eval_batch.reshape((1, look_back, 1))
for i in range(len(test_df)):
    lstm_pred = lstm_model.predict(current_batch)[0]
    lstm_predictions.append(lstm_pred)
    current_batch = np.append(current_batch[:, 1:, :], [[lstm_pred]], axis=1)
lstm_predictions = scaler.inverse_transform(lstm_predictions).flatten()
final_pred_lstm = y_pred_prophet + lstm_predictions

# Ensemble with weights
weight_xgb = 0.4
weight_gb = 0.4
weight_lstm = 0.2
ensemble_pred_weighted = (weight_xgb * final_pred_xgb) + (weight_gb * final_pred_gb) + (weight_lstm * final_pred_lstm)

# Evaluation with RMSE, MAE, MAPE, and SMAPE
def evaluate_model(pred, model_name):
    mae = mean_absolute_error(y_true, pred)
    mape = mean_absolute_percentage_error(y_true, pred)
    rmse = root_mean_squared_error(y_true, pred)
    smape = symmetric_mean_absolute_percentage_error(y_true, pred)
    print(f"{model_name}:")
    print(f"MAE: {mae}")
    print(f"MAPE: {mape}")
    print(f"RMSE: {rmse}")
    print(f"SMAPE: {smape}")

# Prophet evaluation
evaluate_model(y_pred_prophet, "Prophet")

# Other models evaluation
evaluate_model(final_pred_xgb, "Fine-Tuned XGBoost")
evaluate_model(final_pred_gb, "Fine-Tuned Gradient Boosting")
evaluate_model(final_pred_lstm, "Fine-Tuned LSTM")
evaluate_model(ensemble_pred_weighted, "Weighted Ensemble (XGBoost + Gradient Boosting + LSTM)")

# Plotting residuals scatter plots
residuals_xgb = y_true - final_pred_xgb
residuals_gb = y_true - final_pred_gb
residuals_lstm = y_true - final_pred_lstm
residuals_ensemble = y_true - ensemble_pred_weighted

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.scatter(y_true, residuals_xgb, color='blue', alpha=0.5)
plt.title('Residuals for XGBoost')
plt.xlabel('True Values')
plt.ylabel('Residuals')

plt.subplot(2, 2, 2)
plt.scatter(y_true, residuals_gb, color='green', alpha=0.5)
plt.title('Residuals for Gradient Boosting')
plt.xlabel('True Values')
plt.ylabel('Residuals')

plt.subplot(2, 2, 3)
plt.scatter(y_true, residuals_lstm, color='red', alpha=0.5)
plt.title('Residuals for LSTM')
plt.xlabel('True Values')
plt.ylabel('Residuals')

plt.subplot(2, 2, 4)
plt.scatter(y_true, residuals_ensemble, color='orange', alpha=0.5)
plt.title('Residuals for Ensemble')
plt.xlabel('True Values')
plt.ylabel('Residuals')

plt.tight_layout()
plt.show()
```
