```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from scipy import stats, special
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load the dataset
file_path = 'wind_dataset_(2).csv'
df = pd.read_csv(file_path)

# Preprocess data
df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)
y = df['WIND'].values

# Train-test split
train_size = int(len(y) * 0.8)
train, test = y[:train_size], y[train_size:]

# Define evaluation function
def evaluate_model(test, pred):
    mae = mean_absolute_error(test, pred)
    rmse = np.sqrt(mean_squared_error(test, pred))
    smape = 100 * np.mean(2 * np.abs(pred - test) / (np.abs(test) + np.abs(pred)))
    mape = 100 * np.mean(np.abs((test - pred) / test))
    return mae, rmse, smape, mape

# Models list to store results
results = []
residuals_dict = {}

# ARIMA with Box-Cox Transformation
train_boxcox, lam = stats.boxcox(train + 1e-8)
model = ARIMA(train_boxcox, order=(2, 0, 2))
model_fit = model.fit()
pred_boxcox = special.inv_boxcox(model_fit.forecast(steps=len(test)), lam)
results.append(('ARIMA with Box-Cox', *evaluate_model(test, pred_boxcox)))
residuals_dict['ARIMA with Box-Cox'] = model_fit.resid

# SARIMA with Box-Cox Transformation
model = SARIMAX(train_boxcox, order=(2, 1, 1), seasonal_order=(0, 0, 0, 12))
model_fit = model.fit()
pred_sarima_boxcox = special.inv_boxcox(model_fit.forecast(steps=len(test)), lam)
results.append(('SARIMA with Box-Cox', *evaluate_model(test, pred_sarima_boxcox)))
residuals_dict['SARIMA with Box-Cox'] = model_fit.resid

# ARIMA with Log Transformation
train_log = np.log(train + 1)
model = ARIMA(train_log, order=(2, 0, 1))
model_fit = model.fit()
pred_log = np.exp(model_fit.forecast(steps=len(test)))
results.append(('ARIMA with Log', *evaluate_model(test, pred_log)))
residuals_dict['ARIMA with Log'] = model_fit.resid

# SARIMA with Log Transformation
model = SARIMAX(train_log, order=(1, 0, 1), seasonal_order=(1, 0, 1, 12))
model_fit = model.fit()
pred_sarima_log = np.exp(model_fit.forecast(steps=len(test)))
results.append(('SARIMA with Log', *evaluate_model(test, pred_sarima_log)))
residuals_dict['SARIMA with Log'] = model_fit.resid

# Display results in a table
results_df = pd.DataFrame(results, columns=['Model', 'MAE', 'RMSE', 'SMAPE', 'MAPE'])
print(results_df)

# Residual Analysis for all models
plt.figure(figsize=(12, 8))
for i, (model_name, residuals) in enumerate(residuals_dict.items(), start=1):
    plt.subplot(2, 2, i)
    plt.plot(residuals, label=f'{model_name} Residuals')
    plt.title(f'{model_name} Residuals')
    plt.xlabel('Index')  # Use index instead of date
    plt.ylabel('Residuals')
    plt.legend()

plt.tight_layout()
plt.show()
```
